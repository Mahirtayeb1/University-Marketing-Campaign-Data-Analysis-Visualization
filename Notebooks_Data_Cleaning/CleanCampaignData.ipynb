```python
df['Season'] = df['Name'].str.extract(r'(FA\d{2}|SP\d{2}|SU\d{2}|WI\d{2})')
```

```python
df['Campaign_Type'] = df['Name'].apply(
    lambda x: "Call Campaign" if "Call Campaign" in x else "Campaign"
)
```

```python
import re

def extract_region_status(name):
    if "Call Campaign" in name:
        # Split after "Call Campaign:"
        parts = name.split("Call Campaign:")
        if len(parts) > 1:
            text = parts[1].strip()

            # Try to split into region + status using last comma OR last space before known statuses
            match = re.match(r'^(.*)\s+(No Deposit|ANF|Deposit, No I-20|In Progress|Submitted, Incomplete|Deposit No Action|Admit, No Deposit)$', text)
            if match:
                region = match.group(1).strip().rstrip(",")
                campaign_status = match.group(2).strip()
                return region, campaign_status

            # If nothing matches, assume whole thing is region only
            return text, None

    elif "-" in name:
        return None, name.split("-")[-1].strip()
    else:
        return None, name.strip()

# Apply
df[['Region', 'campaign_status']] = df['Name'].apply(lambda x: pd.Series(extract_region_status(x)))
```
